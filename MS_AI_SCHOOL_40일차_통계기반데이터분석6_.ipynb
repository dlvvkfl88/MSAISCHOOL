# 시계열 분해 : 추세, 계절성 ,주기 ,불규칙성 요소 분리하여 시각화

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

### 시계열 데이터 생성


```python
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100)

# 데이터 프레임 
data = pd.DataFrame({
    'Date': dates,
    'Value':values
})

print(data)
```

## 시계열 분해

```python
result = seasonal_decompose(data['Value'], model= 'additive' , period=7) # 시계열 분해

# 분해 결과 시각화
plt.figure(figsize = (10,8))

# 원본 데이터
plt.subplot(4,1,1)
plt.plot(data['Date'], data['Value'])


# 추세 데이터
plt.subplot(4,1,1)
plt.plot(data['Date'],result.trend )


# 계절성 데이터
plt.subplot(4,1,1)
plt.plot(data['Date'], result.seasonal)

# 불규칙성 데이터

plt.subplot(4,1,4)
plt.plot(data['Date'], result.resid)

plt.tight_layout()
plt.show()
```

## 선형보간 -panda를 이용한 선형보간 실습

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시계열 데이터 생성
dates = pd.date_range(start='2023-01-01', periods=10,freq='D')
values= [1,np.nan,3,np.nan,5,6,np.nan,8,9,np.nan]


#데이터프레임 생성
data = pd.DataFrame({
    'Date': dates,
    'Value':values
})

print(data)
```

### 선형 보간 적용


```python
data['Value_linear'] = data['Value'].interpolate(method= 'linear')

print(data)
```

```python
#시계열 데이터 시각화
plt.plot(data['Date'],data['Value'], 'o-', label='Original')
plt.plot(data[ 'Date'], data[ 'Value_linear'], 'x-', label='Linear Interpolation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Linear Interpolation')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
# 시계열 데이터 생성
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)
# 데이터 프레임 생성
data = pd.DataFrame({
    'Date':dates ,
    'Values':values
})
# 결측치 생성
data.loc[100:200, 'Values'] = np.nan


# 결측치 체크 확인
print(data.loc[100:200])
```

## 스플라인 보간법은 유효한 데이터에만 적용 가능 , 그래서 결측치 제거 대처 작업 필요 

결측치 처리(보간법 적용)

```python
data['Value_filed'] = data['Values'].interpolate(method='linear')

print(data.loc[100:200])
```

```python
# 스플라인 보간법 적용 
spl = CubicSpline(data.index, data['Value_filed'])

data['Value_spline'] = spl(data.index)

print(data.loc[100:200])
```

```python
# 시계열 데이터 시각화
plt.plot(data['Date'], data['Values'], 'o-', label='Original')
plt.plot(data['Date'], data['Value_spline'], 'x-', label='Spline Interpolation')
plt.xlabel('Date')
plt.ylabel('Values')
plt.title('Spline Interpolation')
plt.legend()
plt.show()
```

### 스플라인 보간법 -> 연속성 있는 데이터에서 적용됩니다. !! 결측값제거하거나 다른 방법으로 대처 하시고 진행해야 합니다

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시계열 데이터 생성
dates = pd.date_range(start = '2023-01-01', periods=1000, freq='D')
values= np.random.randn(1000)

# 데이터 프레임 생성
data = pd.DataFrame({
    'Date': dates,
    'Value': values
})

# 결측치 생성
data.loc[50:200, 'Value']  = np.nan
print(data.loc[50:200])
```

## pandas 에서 제공하는 최근접 이웃 보간 시각화

```python
# 최근접 이웃 보간 적용
data['Value_filed'] = data['Value'].interpolate(method = 'nearest')
print(data)
```

```python
# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], 'o-', label='Original')
plt.plot(data['Date'], data['Value_filed'], 'x-', label='Nearest Neighbor Interpolation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Nearest Neighbor Interpolation')
plt.legend()
plt.show()
```

# 시계열 데이터에서 결측치를 평균값으로 대처한 후 시각화 실습

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 시계열 데이터 생성
dates = pd.date_range(start = '2023-01-01', periods=1000, freq='D')
values= np.random.randn(1000)

# 데이터 프레임 생성
data = pd.DataFrame({
    'Date': dates,
    'Value': values
})
print(data)
# 결측치 생성
data.loc[50:200, 'Value']  = np.nan
print(data.loc[50:200])
```

## 평균값으로 대처한 후 시각화

```python
mean_value = data['Value'].mean()
print(mean_value)
data['Value_filed'] = data['Value'].fillna(mean_value)
print(data.loc[50:100])

# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], 'o-', label='Original')
plt.plot(data['Date'], data['Value_filed'], 'x-', label='Mean Imputation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Mean Imputation')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
```

```python
# 시계열 데이터 생성
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)

# 데이터 프레임워크 생성

data = pd.DataFrame({
    'Date':dates,
    'Value': values
})

# 결측치 생성
data.loc[50:150, 'Value'] = np.nan
print(data.loc[50:150])
```

## ARIMA 모델 활용하여 결측치 예측과 대처 실습

```python
model = ARIMA(data['Value'].dropna(), order= (1,1,1)) # 모델돌리기위해 결측치제거 order라는 매개변수는 ar 1 차분 1 MA 1 로 설정  
model_fit = model.fit()

# 결측치 예측

predicted_values = model_fit.predict(start=50, end=2000)
print(predicted_values)

# 결측치 대처 
data.loc[50:150, 'value temp'] = predicted_values
print(data[50:150])
```

```python
# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], 'o-', label='Original')
plt.plot(data['Date'], data['value temp'], 'x-', label='ARIMA Imputation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('ARIMA Imputation')
plt.legend()
plt.show()
```
