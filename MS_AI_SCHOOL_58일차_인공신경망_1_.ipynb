```python
def mean_squred_error(y_true, y_pred):
  mean_temp = np.mean((y_true - y_pred)**2)

  return mean_temp


"""
예ㅛㅣ
"""

y_true = np.array([1,2,3,4])
y_pred = np.array([2,3,5,6])

mean_value = mean_squred_error(y_true, y_pred)
print(mean_value)
```

# Binary Cross Entropy Loss 간단한 실습

```python
import numpy as np
```

```python
def binary_cross_entropy_loss(y_true_temp, y_pred_temp):
  epsilon = 1e-5
  n = y_true.shape[0]
  print("n value >>", n)
  loss = -np.mean(y_true_temp * np.log(y_pred_temp + epsilon) + (1-y_true_temp) *np.log(1-y_pred_temp + epsilon) )
  return loss


y_true_temp = np.array([1,0,1,1])
y_pred_temp = np.array([0.99,0.1,0.8,0.7])

loss = binary_cross_entropy_loss(y_true_temp,y_pred_temp)
print(loss)
```

```python
import torch
import torch.nn as nn
```

# 인공신경망 간단하게 실습 100 에포크 마다 산점도 시각화 실습

```python
# 인공신경망 모델 클래스 정의
class SimpleNet(nn.Module):
  """
  입력층 2개 은닉층 1개 (노드 5) 출력층 1개 (씨그모이드 활성화) 이진분류
  """
  def __init__(self):
    super(SimpleNet , self).__init__()
    self.fc1 = nn.Linear(2,5)
    self.fc2 = nn.Linear(5,1)
    pass
  def forward(self,x):
    x = torch.relu(self.fc1(x))
    x = torch.sigmoid(self.fc2(x))

    return x




```

## 데이터 생성

```python
x = torch.randn(100,2)
y = ((x[:,0] > 0) & (x[:,1] > 0)).float() # true false 1, 0 으로 나옴 정답지

print(y)
```

# 모델 선언

```python
model = SimpleNet()

print(model)
```

### Loss 함수량 옵티마이저 정의

```python
# 이진함수 풀거니까
criterion = nn.BCELoss()
optimizer = torch.optim.SGD(model.parameters(), lr= 0.1)
```

## 학습

```python
for epoch in range(1000):

  # 순전파 단계
  y_pred = model(x)

  # loss
  loss = criterion(y_pred.squeeze(),y)

  # 역전파
  optimizer.zero_grad() # 옵티마이저의 모든 가중치대한 변화 0으로 초기화 -> 이전 기울 값이 남아서 초기화해야함
  loss.backward() # 손실함수 역전파 수행
  optimizer.step() # 옵티마이저 가중치 업데이트

  if epoch % 10 ==0:
    plt.clf()
    plt.scatter(x[:,0], x[:,1], c=y_pred.detach().numpy().squeeze())
    plt.show()

print("학습 완료!!!")
```

# 선형 회귀
단순 선형회귀 |

```python
import numpy as np
```

### 예시 데이터 생성

```python
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])
print(x)
print(y)
```

## 기울기와 절편 구하기

```python
beta_1 = np.sum((x-np.mean(x)) * (y - np.mean(y))) / np.sum((x-np.sum((x-np.mean(x))**2)))
beta_0 = np.mean(y) -beta_1 *np.mean(x)
```

```python
# 결과축력
print( "기울기 ", beta_1)
print( "절편 ", beta_0)

#y = 1.0x + 1.0
```

### 단순 선형 회귀의 개념 간단 실습 - 최소 제곱법을 이용한 단순 선형 회귀 구현

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 데이터 생성
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])

# 산점도 그리기
plt.scatter(x,y)
plt.show()
```

```python
# 회귀선 그리기
plt.scatter(x,y,)
plt.plot(x, beta_0 + beta_1 * x , color='red')
plt.show()
```
