# Binary Corss Entropy Loss 간단한 실습

```python
import numpy as np
```

```python
def binary_cross_entropy_loss(y_true_temp, y_pred_temp):
  epsilon =1e-5
  n = y_true_temp.shape[0]
  print("n value >> ",n)

  loss = -np.mean(y_true_temp * np.log(y_pred_temp + epsilon) + (1-y_true_temp) * np.log(1-y_pred_temp + epsilon))

  return loss


y_true_temp = np.array([1,0,1,1])
y_pred_temp = np.array([0.99,0.1, 0.8, 0.7])

loss = binary_cross_entropy_loss(y_true_temp, y_pred_temp)

print(loss)
```

# Cross Entorpy Loss 실습

# # 인공신경망 - Leaky ReLU 함수 실습

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def leaky_relu(x,alpha=0.01):
  return np.where(x >=0, x, alpha)


x = np.linspace(-5,5,100)
y = leaky_relu(x)

plt.plot(x,y)
plt.plot('Leaky RELU FUCNTION')
plt.plot('INPUT')
plt.plot('OUTPUT')
plt.show()
```

# 실습 -06

```python
import torch
import torch.nn as nn
```

```python
# 인공신경망 클래스 정의
class SimpleNet(nn.Module):
  def __init__(self):
    super(SimpleNet, self).__init__()
    self.fc1 = nn.Linear(2,5)
    self.fc2 = nn.Linear(5,1)
    pass

  def forward(self, x ):
    x = torch.relu(self.fc1(x))
    x = torch.sigmoid(self.fc2(x))
    return x

# 데이터를 생성합니다
X = torch.randn(100,2)
y = ((X[:,0]>0) & (X[:,1] >0 )).float()

# 모델 생성
model = SimpleNet()
print(model)
# 손실 함수 옵티마이저 정의
criterion = nn.BCELoss()

optimizer = torch.optim.SGD(model.parameters(), lr= 0.1 )
```

# 학습

```python
for epoch in range(1000):
  # 순전파 단계
  y_pred = model(X)
  print(y_pred)

  # loss
  loss = criterion(y_pred.squeeze(), y)
  # 역전파
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
```

```python
# 시각화
if epoch % 10 ==0:
  plt.clf()
  plt.scatter(X[:,0], X[:,1], c=y_pred.detach().numpy().squeeze())
  plt.show()


print("학습 종료! ")
```


# 단순 선형회귀 개념 간단실습

```python
import numpy as np
```

```python
# 예시데이터 ㅅ행성
x = np.array([1,2,3,4,5])
y = np.array([2,3,4,5,6])

print(x)
print(y)
```

# 기울기와 절편구하기

```python
beta_1 = np.sum(((x-np.mean(x))) * (y - np.mean(y))) / np.sum((x-np.mean(x))**2)

beta_2 = np.mean(y) - beta_1 * np.mean(x)

```

```python
# 결과ㅣ
print("기울기:", beta_1)
print("절편:", beta_2)
```
