```python
# IQR기반 이상치 검사
# 이상치 : Q3 + 1.5 * IQR보다 높거나 Q1 - 1.5 * IQR보다 낮은 값을 의미함
q3 = red_white.quantile(0.75) # df['국어'].quantile(0.75) 처럼 특정 열만 적용 가능
q1 = red_white.quantile(0.25)

q3,q1

iqr =q3 - q1

outlier_up = q3 + 1.5*iqr
outlier_down = q1-1.5*iqr
print(outlier_up)
```

지금 outlier 처리가 많이 걸림 이 에러는 "Can only compare identically-labeled Series objects"라는 메시지로, 같은 레이블을 가진 시리즈 객체들끼리만 비교할 수 있다는 의미입니다. 해당 에러가 발생하는 이유는 'red_white[col]'과 'outlier_up' 모두 시리즈 객체가 아니라서 비교할 수 없기 때문입니다.

해결 방법은 다음과 같이 간단합니다. 'outlier_up'과 'outlier_down'은 각각 스칼라 값이므로 이들과 각 시리즈 열을 비교할 때에는 broadcasting 기능을 사용하여 비교를 수행해야 합니다. broadcasting을 이용하여 각 열의 모든 값을 'outlier_up' 또는 'outlier_down'과 비교할 수 있습니다.

다음은 코드를 수정하여 broadcasting을 사용하는 방법입니다:
여기서 outlier_up[col]와 outlier_down[col]는 각 열에 해당하는 'outlier_up'과 'outlier_down' 값을 broadcasting하여 비교하게 됩니다. 이렇게 하면 각 열의 모든 값을 outlier 기준과 비교할 수 있게 됩니다.
-- 즉 outlier_up[col] 해줘야함 for문 돌때

```python
import numpy as np

# outlier 처리
cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
        'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates',
        'alcohol', 'quality']

for col in cols:
    red_white[col] = np.where(red_white[col] > outlier_up[col], np.nan, red_white[col])
    red_white[col] = np.where(red_white[col] < outlier_down[col], np.nan, red_white[col])
```

```python
# missing value 처리 # IQR 넘어가는 이상치들을 평균으로 대체
import numpy as np

cols = ['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar', 'chlorides',
        'free sulfur dioxide', 'total sulfur dioxide', 'density', 'pH', 'sulphates',
        'alcohol', 'quality']


for col2 in cols:
  red_white[col2] = red_white[col2].fillna(red_white[col2].mean())
```

```python
# Wine만 빼고 DataFrame 생성

feature_df = red_white_scaled[red_white_scaled.columns.difference(['Wine'])]
label_df = red_white_scaled['Wine']

print(feature_df.shape, label_df.shape)
```

## 이제 Pandas로 정제한 값들을 numpy로 변화

```python
feature_np = feature_df.to_numpy().astype('float32')
label_np = label_df.to_numpy().astype('float32')

print(feature_np.shape, label_np.shape) # numpy 배열 형태로 생성
```

1번 데이터넣기전 하기전에 셔플준비

```python
s = np.arange(len(feature_np)) # [ 0~ 6497] 배열형태
np.random.shuffle(s) # 그값 을 섞으면 [1 ,2356. 3513] 이런식

feature_np = feature_np[s] # 섞인 인덱스를 다시 feature_np로 생성
label_np = label_np[s]
```

# train _test 셋 분리

```python
split = 0.15

test_num = int(split*len(label_np)) # 115개 6497개 중 974개 를 test로쓰겠다
print(test_num)

x_test = feature_np[0:test_num] # 974개는 학습용 평가용으로 사용
y_test = label_np[0:test_num]

x_train = feature_np[test_num:] # 5523개는 트레인데이터
y_train = label_np[test_num:]

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

2번단계 model.add w,b 정의

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Flatten, Dense
from tensorflow.keras.optimizers import SGD, Adam
```

# 여기가 w,b 초기화 및 모델층 만드는 정의를 하는곳
이므로 w,b가 단위가 안맞아 W,b 원하는대로 안변하므로 이전에 스케일링이나 정규화 해야함
-위에서 표준화를 했으므로 model.add()수행

```python
model =Sequential()

model.add(Dense(1,input_shape=(feature_np.shape[1],),activation='sigmoid'))
# 분류 sigmoid 0이냐 1이냐 이진분류이므로 출력층 노트1개
# input_shape는 'Wine'컬럼을 제외한 나머지 변수 값인 정답에 필요한 데이터수 인 12개
# activation함수는 0.5기준으로 1과 0을 구분하여 출력층 값을 예측하는 sigmoid함수를 사용
```

## 3번 loss 함수정의 및 옵티마이저  (미분하여 global minimum을 찾는 방법을 정의(경사하강법))  model.compile()

```python
model.compile(optimizer=SGD(learning_rate=1e-3),
              loss = 'binary_crossentropy',
              metrics = ['accuracy'])

# 이진 분류 이므로 loss는 binary_crossentropy 를 적어주고 SGD 옵티마이저하이퍼파라미터 1e-3 임의선택
```

# 만든 모델을 통한 학습 Learning 수행
w,b 최적의값을 찾음 경사하강법을 통해
global minimum

```python
from datetime import datetime

start_time = datetime.now()

hist = model.fit(x_train, y_train, epochs=400, validation_data=(x_test, y_test), verbose=2)

end_time = datetime.now()

print('\nElapsed Time => ', end_time - start_time)
```

```python
model.evaluate(x_test, y_test) # 모델평가 # loss값과 accuracy 0.04로 0에 가깝게 줄음
# accuracy가 상당히 높은 98%임
```

```python
import matplotlib.pyplot as plt

plt.title('Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.grid()

plt.plot(hist.history['loss'], label='train loss')
plt.plot(hist.history['val_loss'], label='validation loss')

plt.legend(loc='best')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.title('Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.grid()

plt.plot(hist.history['accuracy'], label='train accuracy')
plt.plot(hist.history['val_accuracy'], label='validation accuracy')

plt.legend(loc='best')

plt.show()
```
