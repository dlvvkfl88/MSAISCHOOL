```python
import cv2
import numpy as np
```

```python
cap = cv2.VideoCapture('output.avi')

if cap.isOpened() == False:
    print("동영상을 열 수 없습니다.")
    exit(1)

while True:
    ret, img_frame = cap.read()
    if ret == False:
        print( '동영상 읽기 완료')
        break

    cv2.imshow( 'frame', img_frame)
    print( cap.get( cv2.CAP_PROP_POS_FRAMES ) )

    key = cv2.waitKey(30)

    if key== 27:
        break


cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
